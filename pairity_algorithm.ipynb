```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib import rcParams

%matplotlib inline

# figure size in inches optional
rcParams['figure.figsize'] = 11 ,8
```

```python
def show_images(dataset, model, imageA, imageB):
  path_in = '/content/drive/MyDrive/thesis-data/xformer-paired-viz-main/images/' + dataset + '/'
  path_out = '/content/drive/MyDrive/thesis-data/xformer-paired-viz-main/results/' + dataset + '/' + model + '/'

  # read images
  img_1 = mpimg.imread(path_in + imageA + '.png')
  img_2 = mpimg.imread(path_in + imageB + '.png')
  img_12 = mpimg.imread(path_out + imageA + '_simmap_with_' + imageB + '.png')
  img_21 = mpimg.imread(path_out + imageB + '_simmap_with_' + imageA + '.png')

  # display images
  fig, ax = plt.subplots(2,2)
  ax[0][0].imshow(img_1)
  ax[0][1].imshow(img_2)
  ax[1][0].imshow(img_12)
  ax[1][1].imshow(img_21)
```
